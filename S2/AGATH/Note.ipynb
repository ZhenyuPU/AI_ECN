Suppression des strat´egies strictement domin´ees:

关于博弈论（game theory）中消除严格支配策略（strictly dominated strategies）的内容

> - Common Knowledge（共知）： 文中提到了"connaissance commune"，表示玩家对于其他玩家的策略选择有着共同的了解。例如，Alice知道Bob知道她不会选择一个被严格支配的策略，Bob也知道Alice知道他知道，以此类推。
> - 消除严格支配策略： 在博弈论中，如果一种策略在任何情况下都比其他策略更好，那么它就是严格支配的。消除严格支配策略是通过逐步删除这些被严格支配的策略，以简化博弈的过程。这样做不会改变博弈的最优解，因为被删除的策略永远不会是最佳选择。
> - 新的支配关系： 在删除一些策略后，剩下的策略可能会出现新的支配关系。这意味着在删掉一些策略后，可能会有其他策略成为新的最佳选择。

综合起来，这段文字描述了在共同知识的背景下，玩家可以通过消除严格支配策略来简化博弈，从而得到一个更简单但等效的博弈形式。这个过程可以通过逐步删除严格支配策略，并在每一步考虑共同知识的影响。

消除顺序不重要： 消除严格支配策略的顺序不影响最终的结果。不论选择哪一种顺序，最终得到的简化后的博弈形式都是等效的。

仅消除纯策略： 在消除过程中，只考虑纯策略（pure strategies），而不考虑混合策略（mixed strategies）。这是因为如果一个策略不被另一个策略严格支配，那么在混合策略中总存在一个纯策略，可以阻止这种支配关系的发生。

在博弈论中，有时候我们考虑的不仅仅是严格的支配关系，还会引入一些更弱的形式。文中提到的是两种相对较弱的支配形式：**"très faiblement"（非常弱地）** 和 **"faiblement"（弱地）** 的支配。

1. **Très Faiblement（非常弱地）支配：** 一种策略 si' 被称为"très faiblement"支配另一种策略 si，如果对于si的所有其他玩家的策略组合 s-i，都有 $u(si, s-i) \leq u(si', s-i)$。这意味着在任何其他玩家采取的策略组合下，si' 对于si来说都至少不比si差。这是一种相对宽松的支配关系。

2. **Faiblement（弱地）支配：** 一种策略 si' 被称为"faiblement"支配另一种策略 si，如果si' "très faiblement"支配si并且存在其他玩家的策略组合 s-i，使得 \(u(si, s-i) < u(si', s-i)\)。这意味着在某些其他玩家采取的策略组合下，si' 对于si来说至少是更好的选择，而不是仅仅不比si差。

这两种形式的支配关系都是相对于其他玩家的策略组合而言的。它们允许我们更灵活地考虑策略的优劣关系，而不仅仅是在所有可能情况下的严格支配。这样的概念在博弈论中有助于更全面地理解玩家的决策动机和可能的博弈结果。

### Equilibre de Nash

**Meilleure Stratégie**: La stratégie $s_i^*$ du joueur i est une meilleure réponse aux stratégies s_{−i} des autres joueurs s’il n’existe pas une autre stratégie si telle que $ui((si, s−i )) > ui((s_i^*, s−i ))$.

**Equilibre de Nash**: Le profil de stratégies s = (s1, . . . , sn) est un équilibre de Nash si pour tout joueur i, si est une meilleure réponse à s−i
.
